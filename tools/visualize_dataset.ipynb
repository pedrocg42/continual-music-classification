# Dataset Visualization

In this notebook we are going to visualize the music from several music datasets. In this context, we would like to classify the gender of music clips, so we will focus on this task.

```python
# Usual Libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn

# Librosa (the mother of audio files)
import librosa
import librosa.display
import IPython.display as ipd
import warnings
warnings.filterwarnings('ignore')
```

```python
datasets_path = os.getenv("DATASETS_PATH")
```

## GTZN Dataset

In this section, we are going to visualize the data from the GTZN dataset.

```python
gtzn_annotations_path = f"{datasets_path}/GTZAN/features_30_sec.csv"
df = pd.read_csv(gtzn_annotations_path)
paths = df["filename"].to_numpy()
labels = df["label"].to_numpy()
df
```

```python
index = 42
song_path = f"{datasets_path}/GTZAN/genres_original/{labels[index]}/{paths[index]}"
y, sr = librosa.load(song_path)
print('y:', y, '\n')
print('y shape:', np.shape(y), '\n')
print('Sample Rate (KHz):', sr, '\n')

# Verify length of the audio
print('Check Len of Audio:', 661794/22050)
```

```python
plt.figure(figsize = (25, 5))
librosa.display.waveshow(y=y, sr=sr, color="#A300F9")
plt.title("Sound Waves in Reggae 36", fontsize = 23)
```

```python
# Default FFT window size
n_fft = 2048 # FFT window size
hop_length = 512 # number audio of frames between STFT columns (looks like a good default)

# Short-time Fourier transform (STFT)
stft = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_length))
print('Shape of D object:', np.shape(stft))

# Convert an amplitude spectrogram to Decibels-scaled spectrogram.
stft_db = librosa.amplitude_to_db(stft)

# Creating the Spectogram
plt.figure(figsize = (25, 5))
librosa.display.specshow(stft_db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='log', cmap='cool')
plt.colorbar()
```

```python
mel_stft = librosa.feature.melspectrogram(y=y, sr=sr)
mel_stft_db = librosa.amplitude_to_db(mel_stft)
plt.figure(figsize = (25, 5))
librosa.display.specshow(mel_stft_db, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel', cmap='cool')
plt.colorbar()
mel_stft.shape
```
