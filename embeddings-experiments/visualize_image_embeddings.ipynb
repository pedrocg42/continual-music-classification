```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE

model_name = "vit"
```

## CIFAR

```python
train_embeddings = np.load(f"image/{model_name}/cifar100/{model_name}_cifar100_train_embeddings.npy")
train_labels = np.load(f"image/{model_name}/cifar100/{model_name}_cifar100_train_labels.npy")
unique_labels, count_labels = np.unique(train_labels, return_counts=True)
print("Train label count:")
for label, count in zip(unique_labels, count_labels, strict=False):
    print(f"    - {label}: {count}")
```

```python
tsne_train_embeddings = TSNE(n_components=2, learning_rate="auto", init="random", perplexity=30.0).fit_transform(
    train_embeddings
)
```

```python
fig, ax = plt.subplots()
for label in unique_labels:
    ax.scatter(
        tsne_train_embeddings[train_labels == label, 0],
        tsne_train_embeddings[train_labels == label, 1],
        s=6.0,
        label=label,
        edgecolors="none",
    )

ax.grid(True)

plt.show()
```

```python
test_embeddings = np.load(f"image/{model_name}/cifar100/{model_name}_cifar100_test_embeddings.npy")
test_labels = np.load(f"image/{model_name}/cifar100/{model_name}_cifar100_test_labels.npy")
unique_labels, count_labels = np.unique(test_labels, return_counts=True)
print("Validation label count:")
for label, count in zip(unique_labels, count_labels, strict=False):
    print(f"    - {label}: {count}")
```

```python
tsne_test_embeddings = TSNE(n_components=2, learning_rate="auto", init="random", perplexity=30.0).fit_transform(
    test_embeddings
)
```

```python
fig, ax = plt.subplots()
for label in unique_labels:
    ax.scatter(
        tsne_test_embeddings[test_labels == label, 0],
        tsne_test_embeddings[test_labels == label, 1],
        s=6.0,
        label=label,
        edgecolors="none",
    )

# ax.legend()
ax.grid(True)

plt.show()
```
